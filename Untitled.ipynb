```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
cars_df = pd.read_csv("data.csv")
cars_df
```

```python
selected_features = [
    "daysonmarket",
    "horsepower",
    "maximum_seating",
    "mileage",
    "seller_rating",
    "year",
    "price",
    "body_type"
]

cars_analysis = cars_df[selected_features]
cars_analysis
```

```python
correlation_matrix = cars_analysis.corr()
correlation_matrix 
```

```python
y = cars_analysis["price"]
X = cars_analysis.drop(columns=["price","daysonmarket","seller_rating","maximum_seating"])
X = pd.get_dummies(X, columns=['body_type'])
X = X.astype(int)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train
```

```python
from sklearn.metrics import r2_score
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print("R^2:", r2)
```

```python
# Coeficientes de la regresión
coeficientes = model.coef_

# Coeficiente de intercepción
intercepcion = model.intercept_

print("Coeficientes:", coeficientes)
print("Coeficiente de intercepción:", intercepcion)
```

```python
df = pd.DataFrame({'index': range(len(y_test)), 'price': y_test, 'prediction': y_pred})
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['index'][:100], df['price'][:100], label='Real')
plt.plot(df['index'][:100], df['prediction'][:100], label='Predicción')
plt.ylabel('Precio')
plt.title('Comportamiento del precio observado y estimado')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Crea un objeto de escalador
scaler = StandardScaler()

# Ajusta el escalador a los datos de entrenamiento y transforma los datos
X_train_normalized = scaler.fit_transform(X_train)
X_test_normalized = scaler.transform(X_test)
```

```python
model = LinearRegression()
model.fit(X_train_normalized, y_train)
y_pred = model.predict(X_test_normalized)
r2 = r2_score(y_test, y_pred)
print("R^2:", r2)
```

```python
df_normalized  = pd.DataFrame({'index': range(len(y_test)), 'price': y_test, 'prediction': y_pred})
df_normalized 
```

```python
plt.plot(df_normalized['index'], df_normalized['price'], label='Real')
plt.plot(df_normalized['index'], df_normalized['prediction'], label='Predicción')
plt.ylabel('Precio')
plt.title('Comportamiento del precio observado y estimado')
plt.legend()
plt.show()
```

```python
# Coeficientes de la regresión
coeficientes = model.coef_

# Coeficiente de intercepción
intercepcion = model.intercept_

print("Coeficientes:", coeficientes)
print("Coeficiente de intercepción:", intercepcion)
```
